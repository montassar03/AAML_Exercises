```python
import os, sys
```

```python
!wget https://raw.githubusercontent.com/scalable-infrastructure/exercise-2026/main/data/nasa/NASA_access_log_Jul95.gz
!gzip -d NASA_access_log_Jul95.gz
```

```python
!head NASA_access_log_Jul95
```

```python
# Task 1: Which page was called the most?
# cat        → reads the entire log file
# awk        → extracts only the spsecifc column
# sort       → sorts all the pages alphabetically (needed before uniq)
# uniq -c    → counts how many times each page appears
# sort -rn   → sorts by count in descending order (highest first)
# head -n   → shows only the top n most called pages
#----------------------------------------------------------
# cat outputs the entire log file line by line
# Step 1: Extract only the page/URL column
# 'awk' splits each line by spaces and prints column 7
# $7 = the requested URL/page
# Step 2: Sort alphabetically
# 'sort' is required before 'uniq' so identical pages are grouped together
# Step 3: Count occurrences
# 'uniq -c' counts how many times each page appears
# Step 4: Sort by count descending
# 'sort -rn' sorts numerically (-n) in reverse order (-r) → highest count first
# Step 5: Show top 5 results
# 'head -5' keeps only the top 5 most visited pages
print ("Task 1: Which page was called the most?")
!cat NASA_access_log_Jul95 | awk '{print $7}' | sort | uniq -c | sort -rn | head -5
# Get the most visited page (top 1)
print ("Most visited page is:")
!cat NASA_access_log_Jul95 | awk '{print $7}' | sort | uniq -c | sort -rn | head -1

```

```python
# Task 2: What was the most frequent return code?
# Key point: column 9 = the return code
print ("Task 2: What was the most frequent return code?")
print("Most frequent return code is:")
!awk '{print $9}' NASA_access_log_Jul95 | sort | uniq -c | sort -rn | head -1 | awk '{print $2}'
```

```python
# Task 3: How many errors occurred? What is the percentage of errors?
print("Task 3: How many errors occurred? What is the percentage of errors?")

# Count total number of requests
# ! captures shell output as a list in IPython, so we take [0] and convert to int
total = !cat NASA_access_log_Jul95 | wc -l
total = int(total[0])  # convert from list to integer
print("Total requests:", total)

# Count errors (return codes starting with 4 or 5)
errors = !awk '{print $9}' NASA_access_log_Jul95 | grep '^[45]' | wc -l
errors = int(errors[0])  # convert from list to integer
print("Total errors:", errors)

# Calculate percentage
percentage = (errors / total) * 100
print(f"Percentage of errors: {percentage:}%")

# In this exercise, LLMs are used to understand the format of the web server logs. Specifically, they help identify what each column in a log line represents and which metric it corresponds to.
# An LLM can explain that:
# $1  → IP address of the client
# $2, $3 → unused fields
# $4, $5 → timestamp and timezone
# $6 → HTTP method (GET, POST, etc.)
# $7 → the requested URL/page
# $8 → HTTP protocol version
# $9 → HTTP return code (200, 404, 500, etc.)
# $10 → size of the response in bytes

# This allows us to directly extract the right column using awk, making the analysis faster and more accurate.
```
