I was curious to investigate how similarity works and how the model handles typos by replacing a character with one close on the keyboard, adding a letter, or missing one.
The results were surprising; even a single small typo caused the similarity to drop significantly:

- Munich  vs Berlin → 0.739  (correct spelling, high similarity)
- Munich1 vs Berlin → 0.229  (added a number -> similarity collapsed)
- Munish  vs Berlin → 0.148  (one wrong character —> already far from the original)
- Munih   vs Berlin → 0.018  (missing one character —> nearly unrecognizable)
- Munivh  vs Berlin → 0.147  (keyboard-close typo —> similar drop to Munish)
