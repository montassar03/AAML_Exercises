### Creating Rows

Creating `pyspark.sql.Rows` out of your data so you can create DataFrames...

```python
# Convert the CSV into a pyspark.sql.Row
def csv_to_row(line):
    parts = line.split(",")
    row = Row(
      name=parts[0],
      company=parts[1],
      title=parts[2]
    )
    return row

# Apply the function to get rows in an RDD
rows = csv_lines.map(csv_to_row)
```

### Creating DataFrames from RDDs

Using the `RDD.toDF()` method to create a dataframe, registering the `DataFrame` as a temporary table with Spark SQL, and counting the jobs per person using Spark SQL.

```python
# Convert to a pyspark.sql.DataFrame
rows_df = rows.toDF()

# Register the DataFrame for Spark SQL
rows_df.registerTempTable("executives")

# Generate a new DataFrame with SQL using the SparkSession
job_counts = spark.sql("""
SELECT
  name,
  COUNT(*) AS total
  FROM executives
  GROUP BY name
""")
job_counts.show()

# Go back to an RDD
job_counts.rdd.collect()
```

# 4.2–4.7 NASA Dataset

## 4.2
Create a Spark SQL table with fields for IP/Host and Response Code from the NASA log file.

```python
# Exercise 4.2: Create a Spark SQL table with IP/Host and Response Code
print("Task 4.2: Create a Spark SQL table with IP/Host and Response Code")
# Download and extract the NASA log file
!wget "https://github.com/scalable-infrastructure/exercise-2026/blob/main/data/nasa/NASA_access_log_Jul95.gz?raw=true" -O NASA_access_log_Jul95.gz
!gzip -d NASA_access_log_Jul95.gz
# Load the NASA log file
log_lines = sc.textFile("NASA_access_log_Jul95")
# Clean invalid lines same as task 3
valid_lines = log_lines.filter(lambda line: len(line.split()) > 8)
# Extract only the fields we need
# [0]=host [8]=response code
parsed = valid_lines.map(lambda line: (
    line.split()[0],   # host = first column (IP address)
    line.split()[8]    # response_code = 9th column
))
# Convert each tuple into a Spark Row
rows = parsed.map(lambda x: Row(host=x[0], response_code=x[1]))
# Create a DataFrame from the rows
df = spark.createDataFrame(rows)
# Register as a SQL table so we can query it
df.createOrReplaceTempView("nasa_logs")
print("Check table")
df.show(5)
```

## 4.5
Cache the DataFrame and run the same query again. Measure the runtime for caching and for executing the query.

```python
import time
# Task 4.5: Runtime for caching and for executing the query
print("Task 4.5: Runtime for caching and for executing the query")
# Measure runtime without caching
start = time.time()
result_df = df.groupBy("response_code").count().orderBy("count", ascending=False)
result_df.show()
end = time.time()
print("Runtime without caching:", end - start, "seconds")

# Cache the DataFrame and materialize the cache
start_cache = time.time()
df.cache()
df.count()
end_cache = time.time()
print("Caching time:", end_cache - start_cache, "seconds")

# Measure runtime with caching
start_cached_query = time.time()
result_cached_df = df.groupBy("response_code").count().orderBy("count", ascending=False)
result_cached_df.show()
end_cached_query = time.time()
print("Runtime with caching:", end_cached_query - start_cached_query, "seconds")
```

## 4.6 Performance Analysis — Weak Scaling
* Create RDDs with 2x, 4x, 8x, and 16x the size of the NASA log dataset. Persist the dataset in the Spark cache. Use an appropriate number of cores (e.g. 8 or 16).
* Measure and plot the response times for all datasets using a constant number of cores.
* Plot the results.
* Explain the results.


```python
# Task 4.6: Performance Analysis
print("Task 4.5: Runtime for caching and for executing the query")
import time
from pyspark.sql import Row
# Base dataset
rdd_1x = log_lines

# Create larger datasets by union (duplicate data)
rdd_2x = rdd_1x.union(rdd_1x)
rdd_4x = rdd_2x.union(rdd_2x)
rdd_8x = rdd_4x.union(rdd_4x)
rdd_16x = rdd_8x.union(rdd_8x)
# Reuse logic

def create_df(rdd):
    valid = rdd.filter(lambda line: len(line.split()) > 8)
    parsed = valid.map(lambda line: (
        line.split()[0],
        line.split()[8]
    ))
    rows = parsed.map(lambda x: Row(host=x[0], response_code=x[1]))
    return spark.createDataFrame(rows)
# Measure runtime function

def measure_runtime(df):
    start = time.time()

    df.cache()
    df.count()  # materialize

    result = df.groupBy("response_code").count()
    result.collect()

    end = time.time()
    return end - start
# Run experiments
sizes = ["1x", "2x", "4x", "8x", "16x"]
rdds = [rdd_1x, rdd_2x, rdd_4x, rdd_8x, rdd_16x]

times = []

for rdd in rdds:
    df_temp = create_df(rdd)
    runtime = measure_runtime(df_temp)
    times.append(runtime)

for size, t in zip(sizes, times):
    print(f"{size} dataset → {t:.2f} seconds")

# Plot Results
import matplotlib.pyplot as plt

plt.plot(sizes, times, marker='o')
plt.xlabel("Dataset Size")
plt.ylabel("Runtime (seconds)")
plt.title("Scaling Performance")
plt.show()



# Results:
# 1x  dataset → 22.17  seconds
# 2x  dataset → 42.81  seconds
# 4x  dataset → 86.36  seconds
# 8x  dataset → 163.53 seconds
# 16x dataset → 316.07 seconds

# OBSERVATION
# When the dataset doubles, the runtime also roughly doubles:
# 1x  → 22s
# 2x  → 43s  (≈ 2x time)
# 4x  → 86s  (≈ 4x time)
# 8x  → 163s (≈ 8x time)
# 16x → 316s (≈ 16x time)

# EXPLANATION
# This is LINEAR SCALING
# The runtime grows proportionally with the data size
# This happens because the number of cores is constant [2] while only increasing the data size

# WHY IS THIS? (LLM GENERATED)
# With only 2 cores and more data:
# → More data per core to process
# → No extra parallelism to help
# → Runtime increases linearly with data size

# IDEAL WEAK SCALING (LLM generated)
# In an ideal system with weak scaling:
# → If data doubles AND cores double → runtime stays CONSTANT
# → But here cores are fixed → runtime grows linearly
# → This shows Colab's limitation (only 2 cores available)
```
